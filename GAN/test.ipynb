```python
from matplotlib import pyplot as plt
import sys

sys.path.append('../')
sys.path.append('../d2l.py')
import lmy
```

```python
from torchvision import transforms
from torch.utils import data
from torchvision.datasets import FashionMNIST

img_transform = transforms.Compose([
    transforms.ToTensor(),  # 将图像转换为tensor
    transforms.Normalize([0.5], [0.5])  # 归一化
])
mnist_train = FashionMNIST(root='../lmy/data', train=True, transform=img_transform, download=False)
bags_data = []
for i, (img, label) in enumerate(mnist_train):
    if label == 8:
        bags_data.append(img)
train_iter = data.DataLoader(bags_data, 10, shuffle=False, num_workers=4)
```

```python
for bags in train_iter:
    for bag in bags:
        bag = bag.numpy().squeeze()
        plt.imshow(bag, cmap='gray')
        plt.show()
    break
```
